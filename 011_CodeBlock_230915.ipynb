### Deal with missing values

```python
# Replace 0s with NaN
data = data.replace(0, np.nan)
```

```python
pairs = [('NEU1', 'NEU2R'), ('EXT1', 'EXT2R'), ('OPE1', 'OPE2R'), ('AGR1', 'AGR2R'), ('CST1', 'CST2R')]

# Iterate through pairs and apply the transformations
for col1, col2 in pairs:
    # If col1 is missing but col2 is not, apply transformation to col1
    mask1 = data[col1].isna() & data[col2].notna()
    data.loc[mask1, col1] = 7 - data[col2]
    
    # If col2 is missing but col1 is not, apply transformation to col2
    mask2 = data[col2].isna() & data[col1].notna()
    data.loc[mask2, col2] = 7 - data[col1]
    
    # If both are missing, assign random integers to both
    both_missing = data[col1].isna() & data[col2].isna()
    data.loc[both_missing, col1] = np.random.randint(1, 7, size=sum(both_missing))
    data.loc[both_missing, col2] = np.random.randint(1, 7, size=sum(both_missing))

print(data)
```

```python
data['RPR'] = data[['RPR1', 'RPR2', 'RPR3', 'RPR4']].mean(axis=1)

# Define conditions
conditions = [
    (data['RPR'] <= 3), 
    (data['RPR'] > 3) & (data['RPR'] < 4),
    (data['RPR'] >= 4)
]

# Define corresponding choices
choices = [1, 2, 3]

# Create new column using np.select
data['Risk'] = np.select(conditions, choices, default=np.nan)

data.head()
```

```python
# Convert reverse-scored items
cols_to_update = ["NEU2R", "EXT2R", "OPE2R", "CST2R", "AGR2R"]
data[cols_to_update] = 7 - data[cols_to_update]
```

### 3.2. Data augmentation and EDA

### Resampling - Oversampling using SMOTE

```python
from imblearn.over_sampling import SMOTE

# Separate the 'ID' column and the rest of the data
ID_col = data_encoded['ID']
X = data_encoded.drop(columns=['ID', 'Risk'])
y = data_encoded['Risk']

# Apply SMOTE for oversampling
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)

# Create a new DataFrame from the resampled data
resampled_data = pd.DataFrame(X_resampled, columns=X.columns)
resampled_data['Risk'] = y_resampled

# Generate new 'ID' values for the augmented data
last_id_num = int(ID_col.iloc[-1][1:])  # Extract the number part of the last ID 
new_id_nums = range(last_id_num + 1, last_id_num + len(resampled_data) - len(data_encoded) + 1)
new_ids = ['R' + str(num).zfill(3) for num in new_id_nums]

# Add the original IDs and the new IDs to the resampled data
resampled_data['ID'] = list(ID_col) + new_ids

# Reorder the columns to have 'ID' at the front
cols = ['ID'] + [col for col in resampled_data if col != 'ID']
resampled_data = resampled_data[cols]

print(resampled_data)
```

### Bootstrapping and noise injection

```python
n_samples = len(resampled_data)
n_needed = 2000
n_to_bootstrap = n_needed - n_samples

# Bootstrap the needed samples
bootstrap_indices = np.random.choice(resampled_data.index, size=n_to_bootstrap)
bootstrap_samples = resampled_data.loc[bootstrap_indices].copy()

# Function to add noise while maintaining the constraints
def add_noise_and_clip(data, noise_level=0.01):
    numeric_cols = ['NEU1', 'NEU2R', 'EXT1', 'EXT2R', 'OPE1', 'OPE2R', 'AGR1', 'AGR2R', 'CST1', 'CST2R', 
                    'RPR1', 'RPR2', 'RPR3', 'RPR4']
    rpr_cols = ['RPR']
    risk_col = ['Risk']

    # Add noise
    data[numeric_cols] += noise_level * np.random.randn(data[numeric_cols].shape[0], len(numeric_cols))
    data[rpr_cols] += noise_level * np.random.randn(data[rpr_cols].shape[0], 1)
    
    # Clip to maintain the constraints
    data[numeric_cols] = data[numeric_cols].clip(1, 6).round().astype(int)
    data[rpr_cols] = data[rpr_cols].clip(1, 6)
    data[risk_col] = data[risk_col].clip(1, 3).round().astype(int)

    # Ensure categorical columns remain as 0 or 1
    cat_cols = data.drop(columns=numeric_cols + rpr_cols + risk_col).select_dtypes(include=[np.number]).columns
    data[cat_cols] = data[cat_cols].clip(0, 1).astype(int)

    return data

# Apply noise to the bootstrapped samples
noisy_bootstrapped = add_noise_and_clip(bootstrap_samples)

# Concatenate the original data and the noisy bootstrapped data to get the augmented dataset
augmented_data = pd.concat([resampled_data, noisy_bootstrapped], axis=0).reset_index(drop=True)

augmented_data.head()
```

```python
# Categories to compare
active_investors_col = 'Knowledge_Have participated in financial investment.'
other_knowledge_cols = [
    'Knowledge_Aware of financial investment and have intention to invest in the future.',
    'Knowledge_Aware of financial investment but have no intention to participate.',
    'Knowledge_Never heard about financial investment.'
]

# Define a function to get the means for the given knowledge type
def get_means_for_knowledge(data_subset):
    mean_values = {}
    for pair in pairs:
        mean_values[pair] = data_subset[list(pair)].mean(axis=1).mean()
    mean_values['CRTRUN'] = data_subset['CRTRUN'].mean()
    return mean_values

active_data = augmented_data[augmented_data[active_investors_col] == 1]
other_data_mask = augmented_data[other_knowledge_cols].any(axis=1)
other_data = augmented_data[other_data_mask]
active_means = get_means_for_knowledge(active_data)
other_means = get_means_for_knowledge(other_data)

# Plotting
labels = [pair[0] for pair in pairs] + ['CRTRUN']
width = 0.35
x = np.arange(len(labels))

plt.figure(figsize=(12, 6))

plt.bar(x - width/2, [active_means[pair] for pair in pairs] + [active_means['CRTRUN']], 
        width=width, label='Active Investors', color='b')

plt.bar(x + width/2, [other_means[pair] for pair in pairs] + [other_means['CRTRUN']], 
        width=width, label='Others', color='r')

plt.xticks(x, labels)
plt.ylabel('Value')
plt.xlabel('Metrics')
plt.title('Comparison of Means: Active Investors vs Others')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Plotting the risk distribution
risk_labels = ['Seeking', 'Neutral', 'Averse']

active_risk_counts = active_data['Risk'].value_counts().sort_index()
other_risk_counts = other_data['Risk'].value_counts().sort_index()

x_risk = np.arange(len(risk_labels))

plt.figure(figsize=(12, 6))

plt.bar(x_risk - width/2, active_risk_counts, width=width, label='Active Investors', color='b')
plt.bar(x_risk + width/2, other_risk_counts, width=width, label='Others', color='r')

plt.xticks(x_risk, risk_labels)
plt.ylabel('Count')
plt.xlabel('Risk Category')
plt.title('Risk Category Distribution: Active Investors vs Others')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# Calculate the risk percentage for each group
active_risk_percent = active_risk_counts / len(active_data) * 100
other_risk_percent = other_risk_counts / len(other_data) * 100

plt.figure(figsize=(12, 6))

# Plot lines
line1, = plt.plot(x_risk, active_risk_percent, marker='o', label='Active Investors', color='b', linestyle='-')
line2, = plt.plot(x_risk, other_risk_percent, marker='o', label='Others', color='r', linestyle='-')

# Function to annotate line plots
def annotate_lines(points, values):
    for point, value in zip(points, values):
        plt.annotate(
            '{:.2f}%'.format(value),
            (point, value),
            ha='center',
            va='bottom',
            textcoords='offset points',
            xytext=(0,3)
        )

# Annotate the lines
annotate_lines(x_risk, active_risk_percent)
annotate_lines(x_risk, other_risk_percent)

plt.xticks(x_risk, risk_labels)
plt.ylabel('Percentage (%)')
plt.xlabel('Risk Category')
plt.title('Percentage Distribution of Risk Categories: Active Investors vs Others')
plt.legend(loc='upper right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--')
plt.show()
```

## 5. Model Building & Evaluation

```python
# Lists to store the metrics
training_accuracies = []
testing_accuracies = []
f1_scores_train = []
f1_scores_test = []
```

### 5.1. Multinomial logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, f1_score

# Setting up a grid of hyperparameters to search over
param_grid = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],  # Inverse of regularization strength
}

clf_l1 = LogisticRegression(penalty='l1', solver='saga', max_iter=5000)

# Use GridSearchCV with k-fold cross-validation to find the best hyperparameters
clf_cv = GridSearchCV(clf_l1, param_grid, cv=10)  # Using 10-fold CV
clf_cv.fit(X_train, y_train)

# The best hyperparameters from the cross-validation
print(f"Best hyperparameters: {clf_cv.best_params_}")

# Predictions with the best model
y_train_pred = clf_cv.predict(X_train)
y_test_pred = clf_cv.predict(X_test)

# Model Evaluation
train_accuracy_lgr = clf_cv.score(X_train, y_train)
test_accuracy_lgr = clf_cv.score(X_test, y_test)
print(f"Logistic Regression Training Accuracy with L1 regularization: {train_accuracy_lgr:.2f}")
print(f"Logistic Regression Testing Accuracy with L1 regularization: {test_accuracy_lgr:.2f}")

# Additional metrics
print("\nLogistic Regression Training Data Metrics:")
print(confusion_matrix(y_train, y_train_pred))
print(classification_report(y_train, y_train_pred))

print("\nLogistic Regression Testing Data Metrics:")
print(confusion_matrix(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))

# Storing metrics
train_accuracy_lgr = accuracy_score(y_train, y_train_pred)
test_accuracy_lgr = accuracy_score(y_test, y_test_pred)
f1_train_lgr = f1_score(y_train, y_train_pred, average='weighted')
f1_test_lgr = f1_score(y_test, y_test_pred, average='weighted')

training_accuracies.append(train_accuracy_lgr)
testing_accuracies.append(test_accuracy_lgr)
f1_scores_train.append(f1_train_lgr)
f1_scores_test.append(f1_test_lgr)
```
